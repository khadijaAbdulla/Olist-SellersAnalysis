# Data Sprint Project

```python
import pandas as pd 
```

```python
customers = pd.read_csv("olist_customers_dataset.csv")
product_category= pd.read_csv("product_category_name_translation.csv")
sellers= pd.read_csv("olist_sellers_dataset.csv")
products= pd.read_csv("olist_products_dataset.csv")
orders= pd.read_csv("olist_orders_dataset.csv")
order_reviews= pd.read_csv("olist_order_reviews_dataset.csv")
order_payments = pd.read_csv("olist_order_payments_dataset.csv")
order_items= pd.read_csv("olist_order_items_dataset.csv")
geolocation= pd.read_csv("olist_geolocation_dataset.csv")
closed_deals= pd.read_csv("olist_closed_deals_dataset.csv")
marketing= pd.read_csv("olist_marketing_qualified_leads_dataset.csv")
```

## Remove the NULL


```python
products.dropna(subset=['product_category_name','product_description_lenght','product_name_lenght','product_photos_qty','product_weight_g','product_height_cm','product_length_cm'], inplace=True)
```

```python
orders.dropna(subset=['order_approved_at' , 'order_delivered_carrier_date', 'order_delivered_customer_date'], inplace=True)
```

```python
marketing.dropna(subset=['origin'], inplace=True)
```

```python
closed_deals.dropna(subset=['business_segment','lead_type', 'business_type' ], inplace=True)
```

## Translate the Product Category 

```python
products['product_category_name'] = products.merge(
product_category, on='product_category_name', how='left'
)['product_category_name_english'].combine_first(products['product_category_name'])
products
```

## Replace seller id

```python
unique_ids = sellers['seller_id'].unique()
mapping = {old_id: f"Seller{i+1}" for i, old_id in enumerate(unique_ids)}

sellers['seller_name'] = sellers['seller_id'].map(mapping)
```

```python
unique_ids = closed_deals['seller_id'].unique()
mapping = {old_id: f"Seller{i+1}" for i, old_id in enumerate(unique_ids)}

closed_deals['seller_name'] = closed_deals['seller_id'].map(mapping)
```
